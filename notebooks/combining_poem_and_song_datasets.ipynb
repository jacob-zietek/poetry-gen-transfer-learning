```python
song_files2 = glob.glob("../data/kaggle_song_lyrics/*")
```

```python
import os

song_data = [(os.path.basename(f)[:-4], open(f).read().rstrip()) for f in song_files2]
```

```python
song_df2 = pd.DataFrame(song_data, columns=["Artist", "Lyric"])
```

```python
song_df2
```

```python
song_df3 = song_df[["Artist", "Lyric"]]
```

```python
song_df4 = pd.concat([song_df3, song_df2])
```

```python
song_df4.reset_index()
```

```python
poem_df2 = poem_df[["Author", "Content"]]
```

```python
poem_and_songs = pd.concat([poem_df2, song_df4])
```

```python
poem_and_songs.reset_index(drop=True)
```
